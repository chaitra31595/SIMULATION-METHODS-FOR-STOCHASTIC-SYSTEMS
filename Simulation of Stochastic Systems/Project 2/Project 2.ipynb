```python
import networkx as nx
import numpy as np
import math as m
import matplotlib.pyplot as plt
from scipy.stats import *

n=100
p=0.06
Degree=list()
Edge=np.zeros(shape=(n,n))
edgelist=list()
for i in range(0,n):
  for j in range(i+1,n):
    rnd=np.random.uniform(0,1)
    if rnd>=p:
        Edge[i][j]=0
    else:
        Edge[i][j]=1
        e=(i+1,j+1)
        edgelist.append(e)
        

G =nx.Graph()
G.add_nodes_from(range(1,n+1))
G.add_edges_from(edgelist) 
nx.draw(G,with_labels = True) 
Edge1=Edge.transpose()
Final_Edge=Edge+Edge1

#print(Final_Edge)
```

```python
for i in range(0,n):
    sum1=0
    for j in range(0,n):
      sum1=sum1+Final_Edge[i][j]
    Degree.append(sum1)
    
#print(Degree)
Observed_degree=Degree
observed_bin_hist = plt.hist(Degree)[0]
print(Degree)
print(observed_bin_hist)
plt.hist(Degree)
plt.xlabel("Degree of Vertex")
plt.ylabel("Frequency")
plt.title("Histogram of Network Degree Distribution")
#print (observed_bin_hist)
```

```python
Expected_bin_degree=list();
a=np.zeros(shape=(n,n))
for i in range(0,n):
 s=0;
 for j in range (0,n-1):
  rnd=np.random.uniform(0,1)
  if rnd>=p:
   a[i][j]=0;
  else:
   a[i][j]=1;
  s=s+a;
 
a1=a.transpose()
Final_Edge1=a+a1

for i in range(0,n):
    sum1=0
    for j in range(0,n):
      sum1=sum1+Final_Edge1[i][j]
    Expected_bin_degree.append(sum1)


print(Expected_bin_degree)
expected_bin_hist = plt.hist(Expected_bin_degree)[0]
print(expected_bin_hist)
plt.hist(Expected_bin_degree)
plt.xlabel("Degree of Vertex")
plt.ylabel("Frequency")
plt.title("Histogram of Expected Binomial Network Degree Distribution")

```

```python
expected_poisson_degree=np.random.poisson(lam=n*p,size=n)
print(expected_poisson_degree)
expected_poisson_hist = plt.hist(expected_poisson_degree)[0]
print(expected_poisson_hist)
plt.hist(expected_poisson_degree)
plt.xlabel("Degree of Vertex")
plt.ylabel("Frequency")
plt.title("Histogram of Expected Poisson Network Degree Distribution")
```

```python
bin_count=2;
print(observed_bin_hist)
i=0;j=0;
acc_observed_bin_hist=list();
while((i+bin_count)<n):
    s=0;
    for k in range(i,i+bin_count):
      s=s+observed_bin_hist[k];
    acc_observed_bin_hist.append(s)
    j=j+1
    i=i+bin_count

print(acc_observed_bin_hist)


i=0;j=0;
acc_expected_bin_hist=list();
while(i<n):
    s=0;
    for k in range(i,i+bin_count):
      s=s+expected_bin_hist[k];
    acc_expected_bin_hist.append(s)
    j=j+1
    i=i+bin_count

print(acc_expected_bin_hist)


i=0;j=0;
acc_expected_poisson_hist=list();
while(i<n):
    s=0;
    for k in range(i,i+bin_count):
      s=s+expected_poisson_hist[k];
    acc_expected_poisson_hist.append(s)
    j=j+1
    i=i+bin_count

print(acc_expected_poisson_hist)
```
